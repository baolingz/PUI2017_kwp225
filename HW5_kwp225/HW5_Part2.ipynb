# AD and KS test - Gaussian example

I need to figure out what each test function returns. Let me run them once testing an obvious result: is a Gaussian Gaussian? I know the answer so this is a reality check to make sure I understand the output of the functions (and read the manual at the same time!)

I am testing a normal distribution (Gaussian, mean = 0, std = 1) against a Gaussian model, and a Poisson with small $lambda$, or a binomial _w small np against a Gaussian model_. I expect to find that in the first case the Gaussian is, well, Gaussian and in the second case that the binomial with small np is NOT Gaussian

```python
#generate the distribution
alpha = 0.05
np.random.seed(1)
dist_n = np.random.randn(1000) 

#test ad and ks. those are easy
print ("normal on normal", scipy.stats.kstest(dist_n,'norm'))
print ("normal on normal", scipy.stats.anderson(dist_n, dist='norm'))
print ("") 
```

## KS: high p-value pvalue>0.05: H0 that sample comes from a Gaussian distribution cannot be rejected

## AD: at alpha = 0.05 the statistics = 0.26 is smaller than the 0.784 critical value for $alpha$=0.05 (thresh_0.05 = 0.784): the H0 cannot be rejected.

# Testing Poisson vs. Normal

## H0:  The sample Poisson distribution comes from a normal/Gaussian distribution.
### alpha = 0.05

```python
np.random.seed(1)
# poisson with 1000 elements, lambda=100 (high)
dist_p = (np.random.poisson(100, 1000) - 100) * 0.1 #-100 fixes mean, *0.1 fixes stdev

pl.figure()
pl.hist(dist_p, alpha=0.5, label="Poisson")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("similar distributions")
pl.xlabel('x')
pl.ylabel('count')
pl.show()
print("Histograms comparing the Poisson and Normal distributions (mean = 0, standard deviation = 1). " + 
      "They are quite similar.\n")

print ('KS Test: \n', scipy.stats.kstest(dist_p,'norm'),'\n', 
       "P-value = 0.07, which is greater than 0.05. So we can't reject the null hypothesis \n")
print ("Anderson: \n", scipy.stats.anderson(dist_p, dist='norm'),'\n',
       "statistic = 0.61, which is lower than the 0.784 critical value at  alpha=0.05. " +
       "So we can't reject the null hypothesis. \n")

#for KL & chi-square tests
distpdf_n, mybins_n, = np.histogram(dist_n, density=True)
distpdf_p, mybins_p, = np.histogram(dist_p, density=True)

bincenters_n = mybins_n[:-1] + 0.5*(mybins_n[1] - mybins_n[0])
bincenters_p = mybins_p[:-1] + 0.5*(mybins_p[1] - mybins_p[0])


print ("KL Divergence: \n", scipy.stats.entropy(distpdf_p, 
                                scipy.stats.norm.pdf(bincenters_p)),'\n',
       "entropy = 0.01, which is very small. So the difference between the two distributions is very small. \n")
print ("Pearson's Chi-Square: \n", scipy.stats.chisquare(distpdf_p, 
                                                         scipy.stats.norm.pdf(bincenters_p)),'\n',
       "P-value = 1.0. For this test, the null hypothesis is that the categorical data have the same frequency. " + 
       "With such a high p-value, we can't reject the null hypothesis.")
```

# Testing Binomial vs. Normal

## H0:  The sample Binomial distribution comes from a normal/Gaussian distribution.
### alpha = 0.05

```python
#binomial with n=10, p=0.5 => mean=np=5, stdev=sqrt(np(1-p))
n, p = 200, 0.5
dist_b = (np.random.binomial(n, p, 1000) - n * p) / np.sqrt(n * p * (1 - p))  #subtract mean divide by stdev

pl.figure()
pl.hist(dist_b, alpha=0.5, label="Binomial")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("similar distributions")
pl.xlabel('x')
pl.ylabel('count')
pl.show()
print("Histograms comparing the Binomial and Normal distributions (mean = 0, standard deviation= 1). " + 
      "They are similar.\n")

print ('KS Test: \n', scipy.stats.kstest(dist_b,'norm'),'\n', 
       "P-value = 0.02, which is less than 0.05. So we can reject the null hypothesis and determine that the " + 
       "two come from different distributions. \n")
print ("Anderson: \n", scipy.stats.anderson(dist_b, dist='norm'),'\n',
       "statistic = 1.34, which is higher than the 0.784 critical value at  alpha=0.05. " +
       "So we can reject the null hypothesis and determine that the two come from different distributions. \n")

#for KL & chi-square tests
distpdf_b, mybins_b, = np.histogram(dist_b, density=True)

bincenters_b = mybins_b[:-1] + 0.5*(mybins_b[1] - mybins_b[0])


print ("KL Divergence: \n", scipy.stats.entropy(distpdf_b, 
                                scipy.stats.norm.pdf(bincenters_b)),'\n',
       "entropy = 0.009, which is very small. So the difference between the two distributions is very small. \n")
print ("Pearson's Chi-Square: \n", scipy.stats.chisquare(distpdf_b, 
                                                         scipy.stats.norm.pdf(bincenters_b)),'\n',
       "P-value = 1.0. For this test, the null hypothesis is that the categorical data have the same frequency. " + 
       "With such a high p-value, we can't reject the null hypothesis.")
```

# BINOMIAL - INSTRUCTIONS for assignment

with the binomial distribution I can change n*p. larger n*p will make the distribution look more "normal" (i.e. Gaussian). 
In a loop test that the binomial tends to a normal distribution as you increase the value of n*p.

```python
narray = np.linspace(10,1000,100)
ks_s_b = np.zeros(len(narray))
ks_p_b = np.zeros(len(narray))
ad_b   = np.zeros(len(narray))
kl_b   = np.zeros(len(narray))
cs_s_b = np.zeros(len(narray))
cs_p_b = np.zeros(len(narray))

#loop over the different values of n (or n and p)
for i,n in enumerate(narray):
    p = 0.5 #parameter for the binomial, my arbitrary choice

    #generate the distribution
    dist = (np.random.binomial(n, p, 1000) - n * p ) / np.sqrt(n * p * (1-p))
    
    #KS test
    ks_s_b[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_b[i] = scipy.stats.kstest(dist,'norm')[1]

    #anderson
    ad_b[i] = scipy.stats.anderson(dist, dist='norm')[0]
    
    # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist), max(dist), 10) 
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])

    kl_b [i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0] / 1000., 
                                    scipy.stats.norm.pdf(bincenters))
    
    cs_s_b[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0] / 1000., 
                                      scipy.stats.norm.pdf(bincenters))[0]
    cs_p_b[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0] / 1000., 
                                      scipy.stats.norm.pdf(bincenters))[1]
```

```python
threshold = 0.784
fig = pl.figure(figsize = (15,10))
fig.add_subplot(221)
pl.plot(narray, ks_s_b, label='KS statistics')
pl.plot(narray, ks_p_b, label='KS p-value')
pl.xlabel('mean')
pl.ylabel('y')
pl.legend()

fig.add_subplot(222)
pl.plot(narray, ad_b,  label='AD statistics')
pl.plot([narray[0], narray[-1]],[threshold, threshold], label="treshold")
pl.ylim(0,10) #limit the y range or you cannot see the relevant part
pl.xlabel('mean')
pl.ylabel('y')
pl.legend()

fig.add_subplot(223)
pl.plot(narray, kl_b, label='K-L (entropy)')
pl.ylim(0,0.2)
pl.xlabel('mean')
pl.ylabel('y')
pl.legend()

fig.add_subplot(224)
pl.plot(narray, cs_s_b, label='Chi square statistics')
pl.plot(narray, cs_p_b, label='Chi square p-value')
pl.xlabel('mean')
pl.ylabel('y')
pl.legend();
```

#### As the means increase, the distributions approximate normal/Gaussian more.
KS test: Decreasing statistic, increasing p-value, reject null less often.

Anderson: Decreasing statistic, falls below threshold more often, so reject null less often. 

KL divergence: Decreasing entropy, so less separation between the two distributions.

Chi-square: P-value of 1, so can't reject null.

# POISSON - repeat the exercise for the Poisson distribution increasing the lambda parameter of the distribution

```python
ks_s_p = np.zeros(len(narray))
ks_p_p = np.zeros(len(narray))
ad_p   = np.zeros(len(narray))
kl_p   = np.zeros(len(narray))
cs_s_p = np.zeros(len(narray))
cs_p_p = np.zeros(len(narray))

#loop over the different values of n (or n and p)
for i,n in enumerate(narray):
    lam = n #parameter for the binomial, my arbitrary choice

    #generate the distribution
    dist = (np.random.poisson(lam, 1000) - lam) * (1/np.sqrt(lam))
    
    #KS test
    ks_s_p[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_p[i] = scipy.stats.kstest(dist,'norm')[1]
    
    #Anderson
    ad_p[i] = scipy.stats.anderson(dist, dist='norm')[0]
   
    # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist), max(dist), 10) 
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])

    kl_p [i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0] / 1000., 
                                    scipy.stats.norm.pdf(bincenters))
    
    cs_s_p[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0] / 1000., 
                                      scipy.stats.norm.pdf(bincenters))[0]
    cs_p_p[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0] / 1000., 
                                      scipy.stats.norm.pdf(bincenters))[1]
```

```python
fig = pl.figure(figsize = (15,10))
fig.add_subplot(221)
pl.plot(narray, ks_s_p, label='KS statistics')
pl.plot(narray, ks_p_p, label='KS p-value')
pl.xlabel('mean')
pl.ylabel('y')
pl.legend()

fig.add_subplot(222)
pl.plot(narray, ad_p,  label='AD statistics')
pl.plot([narray[0], narray[-1]],[threshold, threshold], label="treshold")
pl.xlabel('mean')
pl.ylabel('y')
pl.ylim(0,10) #limit the y range or you cannot see the relevant part
pl.legend()

fig.add_subplot(223)
pl.plot(narray, kl_p, label='K-L (entropy)')
pl.xlabel('mean')
pl.ylabel('y')
pl.ylim(0,0.1)
pl.legend()

fig.add_subplot(224)
pl.plot(narray, cs_s_p, label='Chi square statistics')
pl.plot(narray, cs_p_p, label='Chi square p-value')
pl.xlabel('mean')
pl.ylabel('y')
pl.legend();
```

Anderson: Decreasing statistic, falls below threshold more often, so reject null less often.
